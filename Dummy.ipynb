```python
import cv2
import matplotlib.pyplot as plt

def display_L_channel_from_rgb(image_path):
    # Load the image in BGR format
    bgr_image = cv2.imread(image_path)
    
    if bgr_image is None:
        print(f"Error: Cannot load image from {image_path}")
        return
    
    # Convert BGR to RGB
    rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
    
    # Convert RGB to LAB
    lab_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2LAB)
    
    # Extract L channel
    L_channel = lab_image[:, :, 0]
    
    # Display the L channel
    plt.imshow(L_channel, cmap='gray')
    plt.title("L Channel (Lightness)")
    plt.axis('off')
    plt.show()
    
    return L_channel  # Optionally return it if needed

# Example usage
image_path = r"C:\Users\gandl\Documents\Python Files\Projects\_Datasets\SAR Image Colorization\agri\s1\ROIs1868_summer_s1_59_p2.png"
L = display_L_channel_from_rgb(image_path)
```
